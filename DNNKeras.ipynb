```python
import pandas as pd
import numpy as np
import keras
data=pd.read_csv("data/applesFull.csv")
X=data.iloc[:, 0:-1].to_numpy();
Y=data.iloc[:, -1].to_numpy();
(n, caracteristicas)=np.shape(X)
```

```python
#crear un modelo de keras
model=keras.Sequential()
#agregar las capas
model.add(keras.layers.Dense(100, activation="relu"))
model.add(keras.layers.Dense(10, activation="relu"))
model.add(keras.layers.Dense(1, activation="sigmoid"))
#compilar el modelo
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
#entrenarlo
hist=model.fit(X, Y, epochs=100, validation_split=.2)
```

```python
pd.DataFrame(hist.history).plot()
```

```python
data=pd.read_csv("mnistTrain.csv")
Y=data.iloc[:, 0].to_numpy()
X=data.iloc[:, 1:].to_numpy()
(n, caracteristicas)=np.shape(X)
print("cantidad de ejemplos",n)
print("cantidad de caracteristicas", caracteristicas)
```

```python
import matplotlib.pyplot as plt
ejemplo=X[50]
label=Y[50]
print("imagen con un", label)
#cambiarle la forma a 28x28
ejemplo=np.reshape(ejemplo, (28, 28))
plt.imshow(ejemplo, cmap="gray")
```

```python
def onehot(Y):
    (n, )=np.shape(Y)
    #crear una matriz de 0s con n ejemplos y 10 salidas
    valores=np.zeros((n, 10))
    #por cada ejemplo, marcar con 1 la salida correcta
    i=0
    for clase in Y:
        valores[i][clase]=1
        i+=1
    return valores

codificado=onehot(Y)
print(codificado)
```

```python
#importar keras
import keras
#crear un modelo secuencial
modelo=keras.Sequential()
#agregar las capas
modelo.add(keras.layers.Dense(50, activation="relu"))
modelo.add(keras.layers.Dense(10, activation="softmax"))#cantidad de salida es igual a la cantidad de clases
#compilar el modelo
modelo.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])
#entrenar el modelo
hist=modelo.fit(X,codificado, epochs=30, validation_split=.2)
```
