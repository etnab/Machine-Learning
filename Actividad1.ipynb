```python
import pandas as pd 
import matplotlib.pyplot as ptl
import seaborn as sns

myData = pd.read_csv('iris.csv')

myData.info #
```

```python
# Id
# SepalLengthCm
# SepalWidthCm
# PetalLengthCm
# PetalWidthCm
# Species

ptl.scatter(myData['PetalLengthCm'], myData['Species'])
ptl.title('Petal Lenght vs Species')
ptl.xlabel('PetalLengthCm')
ptl.ylabel('Species')
```

```python
ptl.scatter(myData['Species'], myData['Id'])
ptl.xlabel('Species')
ptl.ylabel('Id')
```

```python
ptl.scatter(myData['SepalWidthCm'], myData['Species'])
ptl.title('Species vs Sepal Width')
ptl.xlabel('Species')
ptl.ylabel('SepalWidth')
```
