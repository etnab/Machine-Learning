```python
import pandas as pd
myData=pd.read_csv("twitter.csv")
myData.clase=pd.Categorical(myData.clase)
myData.clase=myData.clase.cat.codes
myData.describe()
#clase normal es 1
```

```python
normal=myData.loc[myData["clase"]==1].copy()
anomalos=myData.loc[myData["clase"]==0].copy()
normal.describe()
```

```python
anomalos.describe()
```

```python
import matplotlib.pyplot as plt
plt.boxplot([normal.uniqueURL, anomalos.uniqueURL], labels=["normal", "anomalo"])
plt.show()
```

```python
corr=myData.corr()
import seaborn as sns
sns.heatmap(corr)
print(corr)
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(myData)
plt.show()
```
